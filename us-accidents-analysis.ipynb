### Start Time


```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
sns.histplot(df.Start_Time.dt.hour, bins = 24, stat = "percent")
```

```python
sns.distplot(df.Start_Time.dt.hour, bins = 24 , kde = False , norm_hist = True)
```

- A high percentage of accidents occur between 2 pm to 5 pm

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins = 7 , kde = False , norm_hist = True)
```

### Is the distribution of accidents by hour the same on weekends as on weekday.

```python
weekday_start_time = df.Start_Time[(df.Start_Time.dt.dayofweek != 6) | (df.Start_Time.dt.dayofweek != 5)]
sns.distplot(weekday_start_time.dt.hour, bins = 24 , kde = False , norm_hist = True)
```

```python
weekday_start_time = df.Start_Time[(df.Start_Time.dt.dayofweek != 6) | (df.Start_Time.dt.dayofweek != 5)]
sns.distplot(weekday_start_time.dt.month, bins = 12 , kde = False , norm_hist = True)
```

```python
weekend_start_time = df.Start_Time[(df.Start_Time.dt.dayofweek == 6) | (df.Start_Time.dt.dayofweek == 5)]
sns.distplot(weekend_start_time.dt.hour, bins = 24 , kde = False , norm_hist = True)
```

```python
weekend_start_time = df.Start_Time[(df.Start_Time.dt.dayofweek == 6) | (df.Start_Time.dt.dayofweek == 5)]
sns.distplot(weekend_start_time.dt.month, bins = 12 , kde = False , norm_hist = True)
```
